# Univariate Analysis

## Histogram

```python
features = nums
plt.figure(figsize=(30, 20))
for i in range(0, len(nums)):
    plt.subplot(7, 4, i+1)
    sns.histplot(x=df[features[i]], color='navy')
    plt.xlabel(features[i])
    plt.tight_layout()
```

## Dist Plots

```python
features = nums
plt.figure(figsize=(30, 20))
for i in range(0, len(nums)):
    plt.subplot(7, 4, i+1)
    sns.kdeplot(x=df[features[i]], color='red')
    plt.xlabel(features[i])
    plt.tight_layout()
```

```python
print('Unique Value di kolom ID adalah sebanyak', len(df['ID'].unique()))
print('Unique Value di kolom Recency adalah sebanyak', len(df['Recency'].unique()))
```

#### Pengamatan:
Jika dilihat dari visualisasi distribusi di atas, dapat diketahui bahwa:
<br> 1. Hampir semua kolom tidak ada yang berdistribusi normal
<br> 2. Kolom `Income` , `MntWines` , `MntFruits` , `MntMeatProducts` , `MntFishProducts` , `MntSweetProducts` , `MntGoldProds` , `NumDealsPurchases` , `NumWebPurchases` , `NumCatalogPurchases` , `NumStorePurchases` , `NumWebVisitsMonth` cenderung right-skewed. Kita perlu melakukan sesuatu pada kolom-kolom tersebut pada saat Data Preprocessing 
<br> 3. Dikarenakan pada kolom `Income` terdapat outlier dengan nilai yang sangat besar, maka lebih baik missing value di kolom `Income` bisa diisi dengan median.
<br> 4. Kolom `ID` dan `Recency` memiliki terlalu banyak nilai unik, sehingga bisa dipertimbangkan untuk didrop
<br> 5. Kolom `AcceptedCmp3` , `AcceptedCmp4` , `AcceptedCmp5` , `AcceptedCmp1` , `AcceptedCmp2` , `Complain` , dan `Response` benar bertipe data boolean, sehingga tidak perlu terlalu diperhatikan bentuk distribusinya
<br> 6. Kolom `Year_Birth` benar left-skewed dikarenakan ada outlier yang bernilai data minimum yang kurang masuk akal. Baris dengan nilai `Year_Birth` minimum tersebut bisa dipertimbangkan untuk didrop, agar kolom ini bisa terdistribusi normal

**RATA-RATA Jumlah Pembelian dan Total Belanja Customer**

```python
#Menambahkan kolom total berapa kali pembelian 
df_purchases = df
df_purchases['total_purchases'] = df_purchases['NumDealsPurchases']+ df_purchases['NumWebPurchases'] + df_purchases['NumCatalogPurchases'] + df_purchases['NumStorePurchases']

#Menambahkan kolom total jumlah belanja 
df_purchases['total_amount'] = df_purchases['MntWines'] + df_purchases['MntFruits'] + df_purchases['MntMeatProducts'] + df_purchases['MntFishProducts'] + df_purchases['MntSweetProducts'] + df_purchases['MntGoldProds']

#Menampilkan dataframe
df_purchases.head()
```

```python
fig, axes = plt.subplots(1, # number of rows
                     2, # number of columns
                     figsize=(12, 5), sharex=True, sharey=True)

sns.barplot(x = 'Education', y = 'total_purchases', data = df_purchases, hue = 'Response', ax = axes[0]);
sns.barplot(x = 'Education', y = 'total_purchases', data = df_purchases, ax = axes[1]);
plt.figtext(0,1,'Education VS Total Purchases', size = 'x-large');
```

```python
fig, axes = plt.subplots(1, # number of rows
                     2, # number of columns
                     figsize=(15, 5), sharex=True, sharey=True)
sns.barplot(x = 'Education', y = 'total_amount', data = df_purchases, hue = 'Response', ax = axes[0]);
sns.barplot(x = 'Education', y = 'total_amount', data = df_purchases, ax = axes[1]);
plt.figtext(0,1,'Education VS Total Amount', size = 'x-large');
```

```python
fig, axes = plt.subplots(1, # number of rows
                     2, # number of columns
                     figsize=(15, 5), sharex=True, sharey=True)
sns.barplot(x = 'Marital_Status', y = 'total_purchases', data = df_purchases, hue = 'Response', ax = axes[0]);
sns.barplot(x = 'Marital_Status', y = 'total_purchases', data = df_purchases, ax = axes[1]);
plt.figtext(0,1,'Marital Status VS Total Purchases', size = 'x-large');
```
